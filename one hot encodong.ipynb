```python
import numpy as np 
import pandas as pd
```

```python
df = pd.read_csv("validation_labels.csv")
```

```python
df.head()
```

```python
df = df["labels"]
```

```python
df.head()
```

```python
k = []
m = []
for i in df:
    n = []
    res = str(i)[1:-1]
    a = res.split(',')
    for j in a:
        j = j.replace('u', '')
        j =  j.replace("'", "")
        j = int(j.strip())
        n.append(j)
        if j not in k:
             k.append(j)
    m.append(n)
k = sorted(k)
print(k)
print(len(k))
```

```python
mat = []
for i in m:
    u = [0]* len(k)
    for j in i:
        l = k.index(j)
        u[l] = 1
    mat.append(u)
print(mat)
```

```python
df = pd.DataFrame(mat)
```

```python
df.head()
```

```python
df = df.to_csv("NIT.csv",index = False)
```
